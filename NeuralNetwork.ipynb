```python
import pandas as pd
```

```python
senators = pd.read_csv("final_no_alt_senate_tweets.csv")
senators
```

```python
senators = senators.drop([senators.columns[0], "screen_name", "twitter_full_name", "last_name", "num_tweets_analyzed", "user_id", "name", "avg_neg_sentiment", "avg_neutral_sentiment", "avg_pos_sentiment", "neutral_approval", "neg_approval"], axis=1)
senators
```

stuff to collapse since dependent: `followers_count`, `avg_likes`

```python
senators["avg_likes_per_follower"] = senators["avg_likes"] / senators["followers_count"]

senators
```

```python
senators = senators.drop(["avg_likes", "followers_count"], axis=1)
senators
```

Columns already normalized: `ideology`, `avg_compound_sentiment`

```python
to_normalize = senators.drop(["ideology"], axis=1)
to_normalize
```

```python
means = to_normalize.mean()
sds = to_normalize.std()

to_normalize = to_normalize.apply(lambda x: (x-means)/sds,axis=1)
senators = pd.concat([to_normalize, senators[["ideology"]]], axis=1)

senators
```

```python
X = senators.drop(["pos_approval", "ideology", "extremism"],axis=1)
X
```

```python
y = senators["ideology"]
y = y.apply(lambda x: 0 if x <0.5 else 1)
y 
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.33, random_state=41)
```

```python
def train(X_val, t, permute_test=False, col=None):
    
    if (permute_test):
        X_val2 = X_val.copy()
        X_val2[col] = X_val[col].sample(frac=1, replace=False).values
        X_val = X_val2
        
    X_train, X_test, y_train, y_test = train_test_split(X_val, t, test_size=0.33, random_state=41)
    clf = MLPClassifier(solver='lbfgs', alpha=1e-5, hidden_layer_sizes=(70, 2), random_state=1, max_iter=5000)

    clf.fit(X_train, y_train)
    
    return (1 - abs(clf.predict(X_test) - y_test).sum() / len(X_test))
```

```python
def test_based_importance(X, t, npermutations = 10):
    # initialize what we are going to return
    importances = {}
    for col in X.columns:
        importances[col] = 0
    # find the original accuracy
    orig_accuracy = train(X, t)
    
    # now carray out the feature importance work
    for col in X.columns:
        for perm in range(npermutations):
            
            # get accuracy with shuffled column
            permuted_accuracy = train(X, t, permute_test=True, col=col)
            # print(results)
            importances[col] +=  orig_accuracy - permuted_accuracy
            # subtract to calculate feature importance
        # print(importances[col])
            
        importances[col] = importances[col]/npermutations
    return importances
```

```python
feature_imp = test_based_importance(X, y)
feature_imp
```

```python
import matplotlib.pyplot as plt
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])

plt.bar(feature_imp.keys(), feature_imp.values(), color="#1DA1F2")
plt.xticks(rotation=70)
plt.title("Party Prediction Feature Importance")
```
