```python
import pandas as pd
```

```python
approval_ratings = pd.read_csv("approval.csv")
approval_ratings.drop("Unnamed: 0", axis=1, inplace=True)
```

```python
approval_ratings["last_name"] = approval_ratings["Name"].apply(lambda name: name.split(" ")[1])
approval_ratings
```

```python
ideology = pd.read_csv("ideology3.csv")
ideology["last_name"] = ideology["name"].apply(lambda name: name[2:-1])
ideology = ideology.drop(["name", "rank_from_low", "rank_from_high", "id", "bioguide_id", "state", "district", "percentile"], axis=1)

ideology
```

```python
pd.merge(ideology, approval_ratings, on="last_name", how="inner").to_csv("ideology_approval.csv")
```
